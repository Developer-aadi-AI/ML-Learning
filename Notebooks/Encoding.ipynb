```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    'color' : ['Red', 'Blue', 'Green', 'Blue', 'red'],
    'Shape' : ['Circle', 'Square', 'Triangle', 'Circle', 'Square']
})
```

```python
display(df)
```

# Label Encoding

```python
le_color = LabelEncoder()
le_shape = LabelEncoder()
```

```python
df['Color_encoded'] = le_color.fit_transform(df['color'])
df['Shape_encoded'] = le_shape.fit_transform(df['Shape'])
```

```python
print('\n encoded data \n')
print(df)
```

# one hot Encoding

```python
df_encoded_pandas = pd.get_dummies(df, columns=['color','Shape'])
print('\n One Hot Encoding\n')
print(df_encoded_pandas)
```

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

df = pd.DataFrame({
    'Size' : ['Small', 'Medium', 'Large', 'Medium', 'Small'],
    'Education' : ['Primary', 'Secondary', 'Graduate', 'PostGraduate', 'Graduate']
})
```

```python
print("\n Original Data\n")
print(df)
```

```python
size_order = ['Small', 'Medium', 'Large']
education_order = ['Primary', 'Secondary', 'Graduate', 'PostGraduate']
```

```python
encoder = OrdinalEncoder(categories= [size_order, education_order])
df_encoded = df.copy()
df_encoded[['Size_encoded', 'Education_encoded']] = encoder.fit_transform(df[['Size', 'Education']])
print('\n ordinal Encoded Data')
print(df_encoded)
```
