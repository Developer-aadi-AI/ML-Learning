```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# dataset create
data = {
    'Area': [2100, 1600, 2400, 1416, 3000],
    'Bedrooms': [3, 2, 4, 2, 4],
    'Age': [20, 15, 30, 20, 8],
    'Price': [400, 330, 369, 232, 540]
}

# convert to DataFrame
df = pd.DataFrame(data)

# independent variables (features)
X = df[['Area', 'Bedrooms', 'Age']]

# dependent variable (target)
y = df['Price']
```

```python
# model create
model = LinearRegression()

# train model
model.fit(X, y)
print("Intercept (B0)",model.intercept_)
print("Coefficients(B1,B2,B3)",model.coef_)

# prediction (example)
predicted_price = model.predict([[2500, 3, 10]])

print("Predicted Price in (INR 100000):", predicted_price[0])
```
