```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

```python
X = np.array(
    [[1,2],
     [2,3],
     [3,4],
     [4,5]]
)
y = np.array([4,5,7,10])
```

```python
x_plot = X[:,0].reshape(-1,1)
```

```python
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
y_pred_ols = lin_reg.predict(X)
```

```python
# Ridge regression
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X,y)
y_pred_ridge = ridge_reg.predict(X)
```

```python
# Lasso Regression
lasso_reg = Lasso(alpha = 1.0)
lasso_reg.fit(X,y)
y_pred_lasso = lasso_reg.predict(X)
```

```python
# Print result
print("---Coefficient---")
print(f"OLS Coefficients: {lin_reg.coef_}")
print(f"Ridge Coefficients: {ridge_reg.coef_}")
print(f"Lasso Coefficients: {lasso_reg.coef_}")
```

```python
# Mean Square Error
print("---Mean Squared Error---")
print(f"OLS MSE: {mean_squared_error(y,y_pred_ols)}")
print(f"Ridge MSE: {mean_squared_error(y,y_pred_ridge)}")
print(f"Lasso MSE: {mean_squared_error(y,y_pred_lasso)}")
```

```python
# Visualization
plt.figure(figsize=(8,6))
plt.scatter(x_plot, y, color='black', label='Data Points', s = 60)
plt.plot(x_plot, y_pred_ols, color = 'blue', label='OLS', linewidth=2)
plt.plot(x_plot, y_pred_ridge, color = 'red', label='RIDGE', linewidth=2, linestyle="--")
plt.plot(x_plot, y_pred_lasso, color = 'green', label='LASSO', linewidth=2, linestyle="-.")

plt.title("Lasso Ridge Regression")
plt.xlabel("X1 Feature")
plt.ylabel("Target Y")
plt.legend()
plt.grid(True)
plt.show()
```
