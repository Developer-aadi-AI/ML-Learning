```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
points=np.array([
    [1,2],
    [1,4],
    [1,0],
    [5,2],
    [5,4],
    [6,2]
])

labels=["A","B","C","D","E","F"]
```

```python
ks = [1,2,3,4]
sse = []
for k in ks:
    km = KMeans(n_clusters= k, random_state=42, n_init=50)
    km.fit(points)
    sse.append(km.inertia_)
    print(f"k={k} SSE (Inertia) = {km.inertia_:.6f}")
```

```python
# plot
plt.figure(figsize=(6,4))
plt.plot(ks, sse, '-o')
plt.xticks(ks)
plt.xlabel(' Kmeans ++ (number of clusters)')
plt.ylabel(' SSE/Inertia')
plt.title("Elbow Method")
plt.grid(True)
plt.show()
```

```python
chosen_k = 2
kmeans = KMeans(n_clusters=chosen_k, random_state=42, init = 'k-means++',n_init=10)
kmeans.fit(points)
clusters= kmeans.labels_
centroids = kmeans.cluster_centers_

print("\n Optimal K = ", chosen_k)
for i, label in enumerate(labels):
    print(f"Point {label} {tuple(points[i])} -> cluster {clusters[i]}")
print("Centroids", centroids)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(points[:,0], points[:,1], c=clusters, s=120, cmap='viridis')
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200, c='red', label='centroids')
for i,txt in enumerate(labels):
    plt.annotate(txt, (points[i,0]+0.1, points[i,1]+0.1))
plt.title(f"KMeans with kmeans++ (k={chosen_k})")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.grid(True)
plt.show()
```
