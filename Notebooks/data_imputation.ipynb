```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
data = {
    "id" : [1,2,3,4,5,6],
    'age' : [25,np.nan, 30, 68, np.nan, 23],
    'salary':[50000,60000,np.nan,80000,90000,np.nan],
    'city' : ['New York', 'London',np.nan, 'Paris','London',np.nan ]
}

df = pd.DataFrame(data)
```

```python
display(df)
```

```python
mean_imputer = SimpleImputer(strategy = 'mean')
df_mean = df.copy()

df_mean[['age', 'salary']] = mean_imputer.fit_transform(df_mean[['age', 'salary']])

print('\n Imputed Mean Data')

display(df_mean)
```

```python
median_imputer = SimpleImputer(strategy= 'median')

df_median = df.copy()

df_median[['age', 'salary']] = median_imputer.fit_transform(df_median[['age', 'salary']])
print("\n Imputed Median Data")
display(df_median)
```

```python
mode_imputer = SimpleImputer(strategy = 'most_frequent')
df_data = df.copy()

df_data[['city']] = mode_imputer.fit_transform(df_data[['city']])
print("\n Imputed Mode Data")
display(df_data)
```

```python
df_random = df.copy()

def random_impute(series):
    filled = series.dropna().tolist()
    return series.apply(lambda x: random.choice(filled) if pd.isnull(x) else x)

for col in ['age', 'city', 'salary']:
    df_random[col] = random_impute(df_random[col])

print("\n Random Imputed Data")
display(df_random)
display(df)
    
```
