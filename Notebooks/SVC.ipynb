```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
X = np.array([
    [2,2],
    [4,4],
    [2,0],
    [0,2]
])

y = np.array([1,1,-1,-1])
```

```python
clf = SVC(kernel = 'linear', C = 1e6) # C large value results in hard margin
# Hard margin = the line which can perfectly separate two variables of a dataset, No misclassification allowed and cleaned dataset is used
# soft margin = the line which is unable to perfectly separate two variables of a dataset 
clf.fit(X,y)
```

```python
w = clf.coef_[0]
b = clf.intercept_[0]

support_vector = clf.support_vectors_
support_idx = clf.support_
```

```python
print("w (coefficients): ",w)
print("b (intercepts):",b)
print("Support Vectors:\n", support_vector, "\nSupport Indices:",support_idx)
```

```python
w_norm = np.linalg.norm(w)
dist_to_margin = 1.0/w_norm
full_margin = 2.0/w_norm
print("Half Margin:",dist_to_margin)
print("Full Margin:",full_margin)
```

```python
plt.figure(figsize = (6,6))

plt.scatter(X[y== 1,0],X[y== 1,1], marker='o', label = 'class +1')
plt.scatter(X[y== -1,0],X[y== -1,1], marker='x', label = 'class -1')

plt.scatter(support_vector[:,0], support_vector[:,1], s=200, facecolor = 'none',
             edgecolors = 'k', linewidths=2, label = 'Support Vectors')

xx = np.linspace(-1,6,200)

yy_decision = -(w[0]*xx+b)/w[1]
yy_margin_pos = -(w[0]*xx+b-1)/w[1]
yy_margin_neg = -(w[0]*xx+b+1)/w[1]

plt.plot(xx, yy_decision, linestyle = '-', linewidth = 2, label = 'decision boundary')
plt.plot(xx, yy_margin_pos, linestyle = '--', linewidth = 1, label = 'margin (+1)')
plt.plot(xx, yy_margin_neg, linestyle = '--', linewidth = 1, label = 'margin (-1)')

plt.xlim(-1,6)
plt.ylim(-1,6)
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()

plt.show()
```
